```python
m_unique=[]
for m in range(3,13):
    month_df = df[df['Month'] == m]
    monthuniqueusers = month_df['id'].value_counts()
    m_unique.append(len(monthuniqueusers))
m_unique
m_unique= np.array(m_unique)
```

```python
import matplotlib.pyplot as plt 
import matplotlib


#plt.title('Month v/s Active Users', fontsize="30")
plt.xlabel('Month',fontsize="30")
plt.ylabel('Active Users',fontsize="30")
###### Important ==>

colors = cm.viridis(m_unique / float(max(m_unique)))


plt.bar(['March','April','May','June','July','August','September','October','November','December'],m_unique,color=colors)
plt.xticks(rotation = 90)
#plt.grid()
matplotlib.rcParams['figure.figsize'] = [10,5]
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.show()
```
